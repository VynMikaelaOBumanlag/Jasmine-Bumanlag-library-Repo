```python
import requests
import json

url = "https://raw.githubusercontent.com/VynMikaelaOBumanlag/Jasmine-Bumanlag-library-Repo/refs/heads/main/books.json"

response = requests.get(url)
data = response.json()

print("JSON file loaded")
print(data)
```

```python
ref = db.reference("books")

for book in data["books"]:
    ref.child(str(book["id"])).set(book)

print("Data uploaded successfully!")
```

```python
books_ref = db.reference("books")
books_data = books_ref.get()

print(books_data)
```
